```python
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = None 
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock


class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, selected_items):
        # Logic to place a new order
        order = {"items": selected_items, "total_price": sum(item.price for item in selected_items)}
        self.orders.append(order)


class Restaurant:
    def __init__(self):
        self.food_menu = []
        self.users = []

    def add_food_item(self, food_item):
        # to add a new food item
        food_item.food_id = len(self.food_menu) + 1
        self.food_menu.append(food_item)

    def edit_food_item(self, food_id, new_details):
        # to edit a food item
        food_item = next(item for item in self.food_menu if item.food_id == food_id)
        for key, value in new_details.items():
            setattr(food_item, key, value)

    def remove_food_item(self, food_id):
        # to remove a food item
        self.food_menu = [item for item in self.food_menu if item.food_id != food_id]

    def get_food_list(self):
        # Return the list of all food items
        return self.food_menu


# Example

restaurant = Restaurant()

# Admin functionalities
food1 = FoodItem("Tandoori Chicken", "4 pieces", 240, 0, 50)
restaurant.add_food_item(food1)

food2 = FoodItem("Vegan Burger", "1 piece", 320, 0, 30)
restaurant.add_food_item(food2)

food3 = FoodItem("Truffle Cake", "500gm", 900, 0, 20)
restaurant.add_food_item(food3)

# User functionalities
user1 = User("Amit", "123456", "amit@example.com", "123 Main St", "password123")
restaurant.users.append(user1)

# Assume user logs in
current_user = restaurant.users[0]

# Place New Order
food_list = restaurant.get_food_list()
print("Available Food Items:")
for i, food_item in enumerate(food_list, start=1):
    print(f"{i}. {food_item.name} ({food_item.quantity}) [INR {food_item.price}]")

selected_items = [food_list[int(i) - 1] for i in input("Enter numbers separated by commas to select items: ").split(",")]

current_user.place_order(selected_items)

# Order History
print("Order History:")
for i, order in enumerate(current_user.orders, start=1):
    print(f"{i}. Items: {[item.name for item in order['items']]}, Total Price: {order['total_price']}")

# Update Profile
new_address = input("Enter your new address: ")
current_user.address = new_address
print("Profile updated successfully!")
```
